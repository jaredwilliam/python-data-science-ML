# Principle Component Analysis

It is an unsupervised statistical technique used to examine the interrelations among a set of variables in order to identify the underlying structure of those variables. It is also known sometimes as a general **factor analysis**. 

Where regression determines a line of best fit to a data set, factor analysis determines several orthogonal lines of best fit to the data set. 

Let's use linear data with two dimensions as an example, as in the image below.

There is some data plotted along two features, x and y. We add in a regression line of best fit. We can then add in an orthogonal line. Now we can begin to understand the components. 

Components are a linear transformation that chooses a variable system for the data set such that the greatest variance of the data set comes to lie on the first axis, the second greatest variance on the second axis, and so on. This process allows us to reduce the number of variables used in an analysis. 

So we can see that the first line explains 70% of the variation in the data, where as the second line explains 28%. We are missing 2%.

Note that the components are uncorrelated, since in the sample space they are orthogonal to each other.

If we use this techniques on a data set with a large number of variables, we can compress the amount of explained variation to just a few components. The most challenging part of PCA is interpreting the components. 

When doing PCA, you are going to want to standardize the data as well. 

# PCA with Python

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
sns.set_style('whitegrid')
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df = pd.DataFrame(cancer['data'], columns = cancer['feature_names'])
```

```python
df.head()
```

We have data with 30 dimensions. We really can't visualize that, so we are going to use PCA to get it down to two. 

First we are going to scale our data so it all has single unit variance. 

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
# PCA
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

We have now standardized the data and scaled it down to two components

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c=cancer['target'], cmap='plasma')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

We can see in this plot that we have a very clear separation between benign and malignant tumors by only using two components out of 30. 

So now what do we mean by First Principal Component and Second Principal Component? They don't related 1:1 to a feature in the data, the components are really combinations of the features. 

```python
pca.components_
```

In this numpy array of the components, each row represents a principal component, and each column relates back to the features. We can visualize this relationship with a heatmap. 

```python
# Here we have a dataframe where the rows are the components and the 
# columns are the features
df_comp = pd.DataFrame(pca.components_, columns=cancer['feature_names'])
```

```python
df_comp
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp, cmap='plasma')
```
